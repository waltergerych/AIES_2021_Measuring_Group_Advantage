```python
def eval_rpar0(ranks, rank_len, filename):
    ms=[]
    for r in rs:
        print("Group Size: ", r)
        for p in ps:
            for g in ranks[r][p]:
                groups = [0 if x in g else 1 for x in np.arange(rank_len)]
    #             print(groups)
                e0,e1 = rank_parity(np.arange(rank_len), groups)
                R_par = e0 
                ms.append([p, r, R_par])
    sr = pd.DataFrame(ms, columns=["p","r","R_par0"])
    srg = sr.groupby(by=["p","r"]).mean().unstack()
    srg.columns=[200,500,800]
    ax = srg.plot(style=['^-', 's:', 'o--', '*-'], ms=12)
    ax.set_ylim((0.0,1.0))
    ax.set_yticklabels([])
    ax.set_xlabel(r'$\alpha$')
    ax.get_legend().remove()
    plt.savefig("figures/charts/"+filename+".png", bbox_inches='tight')
    plt.savefig("figures/charts/"+filename+".pdf", bbox_inches='tight')
    plt.show()
```

```python
def eval_rpar(ranks, rank_len, filename):
    ms=[]
    for r in rs:
        print("Group Size: ", r)
        for p in ps:
            for g in ranks[r][p]:
                groups = [0 if x in g else 1 for x in np.arange(rank_len)]
    #             print(groups)
                e0,e1 = rank_parity(np.arange(rank_len), groups)
                R_par = abs(e0 - e1) 
                ms.append([p, r, R_par])
    sr = pd.DataFrame(ms, columns=["p","r","R_par"])
    srg = sr.groupby(by=["p","r"]).mean().unstack()
    srg.columns=[200,500,800]
    ax = srg.plot(style=['^-', 's:', 'o--', '*-'], ms=12)
    ax.set_ylim((0.0,1.0))
    #ax.set_title("Rpar")
    ax.set_yticklabels([])
    ax.set_xticklabels(np.repeat("       ", len(ps)))
    ax.set_xlabel('')
    ax.get_legend().remove()
    plt.savefig("figures/charts/"+filename+".png", bbox_inches='tight')
    plt.savefig("figures/charts/"+filename+".pdf", bbox_inches='tight')
    plt.show()
    
def eval_expDCG(ranks, rank_len, filename):
    ms=[]
    for r in rs:
        print("Group Size: ", r)
        norm = getExpDCGNormalizer(1000, r)
        p_len = r*rank_len
        for p in ps:
            for g in ranks[r][p]:
                dcg = calculaterExpDCG(np.arange(rank_len),g,rank_len,p_len)
                ms.append([p, r, (dcg - norm[0])/(norm[1] - norm[0])])
    sr = pd.DataFrame(ms, columns=["p","r","DCG"])

    srg = sr.groupby(by=["p","r"]).mean().unstack()
    srg.columns=[200,500,800]
    ax = srg.plot(style=['^-', 's:', 'o--', '*-'], ms=12)
    #ax.set_title("ExpDCG")
#     ax.set_ylabel("Fairness Score")
    ax.set_ylabel("Fairness Score")
    ax.set_xticklabels(np.repeat("       ", len(ps)))
    ax.set_xlabel('')
    ax.get_legend().remove()
    plt.savefig("figures/charts/"+filename+".png", bbox_inches='tight')
    plt.savefig("figures/charts/"+filename+".pdf", bbox_inches='tight')
    plt.show()

def eval_expRR(ranks, rank_len, filename):
    ms=[]
    for r in rs:
        print("Group Size: ", r)
        norm = getExpRRNormalizer(1000, r)
        p_len = r*rank_len
        for p in ps:
            for g in ranks[r][p]:
                rr = calculaterExpRR(np.arange(rank_len),g,rank_len,p_len)
                ms.append([p, r, (rr - norm[0])/(norm[1] - norm[0])])
    sr = pd.DataFrame(ms, columns=["p","r","RR"])

    srg = sr.groupby(by=["p","r"]).mean().unstack()
    srg.columns=[200,500,800]
    ax = srg.plot(style=['^-', 's:', 'o--', '*-'], ms=12)
    #ax.set_title("RR")
#     ax.set_ylabel("Fairness Score")
    ax.set_yticklabels("")
    ax.set_xticklabels(np.repeat("       ", len(ps)))
    ax.set_xlabel('')
    ax.get_legend().remove()
    plt.savefig("figures/charts/"+filename+".png", bbox_inches='tight')
    plt.savefig("figures/charts/"+filename+".pdf", bbox_inches='tight')
    plt.show()
```

```python
def eval_padded_yang_metric(pad_ranks, rank_len, window, ks, metric, filename,xs=True,ys=True,leg=True):
    pad_stats=[]
    rank = np.arange(rank_len)
    for r in rs:
#         print("Group Size: ", r)
        size=int(rank_len*r)
        max_m = getNormalizer(rank_len,size,metric,window)
        for k in ks:
            if rank_len*(1-r) >= k:
                for g in pad_ranks[r][k]:
                    gf=calculateNDFairness(rank, g, window, metric, max_m)
                    pad_stats.append([k, r, gf])
    sr = pd.DataFrame(pad_stats, columns=["k","r",metric])

    srg = sr.groupby(by=["k","r"]).mean().unstack()
    srg.columns=[200,500,800]
    ax = srg.plot(style=['^-', 's:', 'o--', '*-'], ms=12)
    #ax.set_title(metric)
    ax.set_ylabel("Fairness Score")
    ax.set_ylim((0.0,1.0))
    if ys:
        ax.set_ylabel("Fairness Score")
    else:
        ax.set_yticklabels([])
    if xs:
        ax.set_xlabel(r'$\alpha$')
    else:
        ax.set_xticklabels(np.repeat("       ", len(ps)))
        ax.set_xlabel('')
    if not leg:
        ax.get_legend().remove()
    plt.savefig("figures/charts/"+filename+".png", bbox_inches='tight')
    plt.savefig("figures/charts/"+filename+".pdf", bbox_inches='tight')
    plt.show()
```
