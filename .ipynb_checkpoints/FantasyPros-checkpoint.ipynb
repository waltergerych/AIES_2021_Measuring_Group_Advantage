**load data from pickle files**

```python
#top 25 experts' rankings of Wide recievers for 16 weeks of the season
# top25_dfs is a dictionary containing dataframes of rankings for one week.
# each column is a ranking of 55 players. The last column "Division" gives the group labels for each player

pkl_file = open('top25_dfs.pickle', 'rb')
top25_dfs = pickle.load(pkl_file)
pkl_file.close()
```

```python
top25_dfs[1]
```

```python
#compute parity scores for all ranks

df_par=pd.DataFrame(index=range(1,17))
experts = list(top25_dfs[1].columns.values)
for exp in experts:
    pars = []
    for week in range(1,17):
        ranks = top25_dfs[week]
        r = ranks[exp]
        g = ranks['Division']
        a0,a1 = rank_parity(r,g)
        pars.append(abs(a1-a0))
    df_par[exp]=pars
```

**Visualize rankings for each week**

```python
# each week has the rankings of players that all experts ranked that week. 
# So they have different number of players each week 

weeks = list(range(1,17,2))
plt.rcParams['figure.figsize'] = 15,6
fig, axs = plt.subplots(1, len(weeks), sharey=True)

i = 0
for week in weeks:
    r = []
    for c in top25_dfs[week].drop('Division', axis=1):
        r.append(top25_dfs[week].sort_values(c)['Division'].values)
    axs[i].imshow(np.transpose(r), cmap='hot', interpolation='nearest')
    axs[i].set_title("Week "+str(week))
    i+=1
plt.show()
```
